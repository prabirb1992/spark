```python
edges=sc.textFile(edgefile)
edges.take(10)
```

```python
def edgeFilter(row):
    if '*' in row or '"' in row:
        return False
    else:
        return True

edgesOnly=edges.filter(edgeFilter).map(lambda x:x.split()).map(lambda x :(int(x[0]),len(x[1:])))
edgesOnly.take(100)
```

```python
edge=edgesOnly.reduceByKey(lambda x,y:x+y)
```

```python
edge.sortBy(lambda x:x[0]).collect()
```

```python
def charparse(row):
    row=row.split(":")
    return (int((row[0].split())[1]),row[1].strip())
charslookup=chars.map(charparse).collectAsMap()
type(charslookup)

```

```python
mostInfluencer=edge.map(lambda x:(charslookup[x[0]],x[1])).sortBy(lambda x: -x[1])
```
